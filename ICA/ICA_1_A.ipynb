    <p>Quickly (about 5 minutes) read the two codes below.</p>

I recommend keeping a code like this where you can find it so that you can easily paste it at the top of your Python code and take control over your visualizations!  </li>

```python
import matplotlib.pyplot as plt
import numpy as np

# Customize rcParams
plt.rcParams.update({
    # Remove top and right spines
    'axes.spines.top': False,
    'axes.spines.right': False,
    
    # Increase font size for axis labels
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    
    # Add a light gray grid with low alpha
    'axes.grid': True,
    'grid.color': 'gray',
    'grid.alpha': 0.1,
    
    # Increase the default figure size
    'figure.figsize': [10, 6],
    
    # Use a higher DPI for better quality
    'figure.dpi': 120,
    
    # Set a global font family
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans', 'Liberation Sans', 'Bitstream Vera Sans', 'sans-serif'],
    
    # Increase the linewidth of the axes
    'axes.linewidth': 1.5,
    
    # Customize the tick marks
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.major.width': 1.5,
    'ytick.major.width': 1.5,
    
    # Use LaTeX for math rendering (if available)
    'text.usetex': False,  # Set to True if LaTeX is installed
    
    # Adjust padding to prevent clipping of tick-labels
    'axes.autolimit_mode': 'round_numbers',
    'axes.xmargin': 0.1,
    'axes.ymargin': 0.1,
    
    # Use a color-blind friendly palette
    'axes.prop_cycle': plt.cycler('color', ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB'])
})

# Example plots
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()

# Plot sine waves with different frequencies
frequencies = [1, 2, 3, 4]
for freq in frequencies:
    y = np.sin(freq * x)
    ax.plot(x, y, label=f'sin({freq}x)')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sine Waves at Different Frequencies', fontsize=16)
ax.legend()

plt.tight_layout()
plt.show()

# Example of a more complex plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5))

# Upper plot: sine and cosine
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.plot(x, np.cos(x), label='cos(x)')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Sine and Cosine', fontsize=18)
ax1.legend()

# Lower plot: exponential decay
ax2.plot(x, np.exp(-x), label='exp(-x)')
ax2.plot(x, np.exp(-0.5*x), label='exp(-0.5x)')
ax2.plot(x, np.exp(-0.2*x), label='exp(-0.2x)')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_title('Exponential Decay', fontsize=14)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE

# Load the wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# Define the two feature sets (corrected indices)
potentially_harmful_features = [0, 1, 2, 3, 4, 7]  # Indices of potentially harmful or neutral features

# Function to perform t-SNE and plot results
def plot_tsne(X, y, features, title):
    # Select and scale the features
    X_selected = StandardScaler().fit_transform(X[:, features])
    
    # Perform t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    X_tsne = tsne.fit_transform(X_selected)
    
    # Plot the results
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
    plt.title(title)
    plt.xlabel('t-SNE feature 1')
    plt.ylabel('t-SNE feature 2')
    plt.colorbar(scatter, label='Wine Class')
    plt.show()

print("\nPotentially Harmful or Neutral Features:")
for idx in potentially_harmful_features:
    print(f"- {wine.feature_names[idx]}")

# Plot t-SNE for potentially harmful features
plot_tsne(X, y, potentially_harmful_features, 't-SNE of Wine Dataset: Potentially Harmful Features')

```
